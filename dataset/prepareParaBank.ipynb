```python
def para(row, anchor_col="sentence1", para_col="sentence2", label_col="label", path="parabank", name="small_diverse", split="train"):
  out_row = {
             "sentence1": row[anchor_col],
             "sentence2": row[para_col],
             "label": 1,
             "path": path,
             "name": name,
             "split": split}
  return out_row
```

```python
import pandas as pd

df = pd.read_csv("raw/parabank/parabank-1.0-small-diverse/parabank.5m.tsv", names=["sentence1", "sentence2"], sep='\t|\n', engine='python', error_bad_lines=True, nrows=55000, encoding="utf-8")
```

```python

out_train = []
out_val = []

max_idx = df.shape[0]
for idx, row in df.iterrows():
    if idx < 50000:
        out_train.append(para(row, split="train"))
    else:
        out_val.append(para(row, split="validation"))


pd.DataFrame(out_train).reset_index().to_csv("raw/out_parabank_small_diverse_train.csv", index=False)
pd.DataFrame(out_val).reset_index().to_csv("raw/out_parabank_small_diverse_validation.csv", index=False)
#pd.DataFrame(out_test).reset_index().to_csv("raw/out_parabank_small_diverse_test_sml.csv")
```
